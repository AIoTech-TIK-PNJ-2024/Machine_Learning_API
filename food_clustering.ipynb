```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Membaca data dari file CSV
data = pd.read_csv('nutrition.csv')

# Mengambil kolom yang ingin digunakan untuk clustering (misalnya kolom 'Feature 1' dan 'Feature 2')
X = data[['calories', 'proteins']].values

# Inisialisasi model KMeans dengan jumlah cluster yang diinginkan
kmeans = KMeans(n_clusters=3)

# Melatih model pada data
kmeans.fit(X)

# Mendapatkan label cluster untuk setiap titik data
y_kmeans = kmeans.predict(X)

# Mendapatkan posisi pusat dari setiap cluster
centers = kmeans.cluster_centers_

# Visualisasi hasil clustering
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hasil Clustering dengan KMeans')
plt.show()
```

```python
labels = kmeans.labels_
data['Cluster'] = labels
```

```python
for cluster_label in np.unique(labels):
    cluster_data = data[data['Cluster'] == cluster_label]
    min_values = cluster_data[['calories', 'proteins']].min()
    max_values = cluster_data[['calories', 'proteins']].max()
    print(f"Cluster {cluster_label}:")
    print(f"   Minimum values: {min_values}")
    print(f"   Maximum values: {max_values}")
    print()
```
